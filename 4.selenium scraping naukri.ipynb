```python
url="https://www.naukri.com/data-science-jobs"
# resp=requests.get(url,headers=headers).text
# print(resp)
```

```python
driver=webdriver.Firefox()
```

```python
driver.get(url)
```

```python
html=driver.page_source
```

```python
print(html)
```

```python
import re
```

```python
re.findall('span.*class="salary"',html)
```

```python
html_parsable=BeautifulSoup(html,'html.parser')
```

```python
html_parsable.title
```

```python
title=[]
salary=[]
exp=[]
company=[]
loc=[]
for i in html_parsable.find_all("a",class_="title fw500 ellipsis"):
    title.append(i.text)
for i in html_parsable.find_all("li",class_="fleft grey-text br2 placeHolderLi salary"):
    salary.append(i.text)
for i in html_parsable.find_all("li",class_="fleft grey-text br2 placeHolderLi experience"):
    exp.append(i.text)
for i in html_parsable.find_all("a",class_="subTitle ellipsis fleft"):
    company.append(i.text)
for i in html_parsable.find_all("li",class_="fleft grey-text br2 placeHolderLi location"):
    loc.append(i.text)
```

```python
len(title),len(salary),len(exp),len(company),len(loc)
```

```python
import pandas as pd
```

```python
table=pd.DataFrame([title,company,salary,loc,exp]).T
```
